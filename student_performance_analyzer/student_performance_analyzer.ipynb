# Problem 2: Student Performance Analyzer 🎓



students = [
    ("Rahul", "Math", 85),

    ("Anu", "Math", 92),

    ("Kiran", "Math", 78),

    ("Rahul", "Science", 88),

    ("Anu", "Science", 95),

    ("Kiran", "Science", 82),

    ("Rahul", "English", 75),

    ("Anu", "English", 89),

    ("Kiran", "English", 91)
]

Each tuple contains:

(student_name, subject, marks)

```python
# creating given data
students = [
    ("Rahul", "Math", 85),
    ("Anu", "Math", 92),
    ("Kiran", "Math", 78),
    ("Rahul", "Science", 88),
    ("Anu", "Science", 95),
    ("Kiran", "Science", 82),
    ("Rahul", "English", 75),
    ("Anu", "English", 89),
    ("Kiran", "English", 91)
]
print(students)
```

```python
# creating list to each one
stu_name=[]
sub_name=[]
marks=[]
for stu , sub , mark in students:
  stu_name.append(stu),sub_name.append(sub),marks.append(mark)
print(stu_name)
print(sub_name)
print(marks)
```

## 1. Calculate Total Marks of Each Student

Rahul -> 248

Anu -> 276

Kiran -> 251

```python
stud_total_marks={}
for name , mark in zip(stu_name , marks):
  if name in stud_total_marks:
    stud_total_marks[name]=stud_total_marks[name]+mark
  else:
    stud_total_marks[name]=mark
for name in stud_total_marks:
  print(f'{name} ->{stud_total_marks[name]}')
```

##2. Calculate Average Marks of Each Student

Expected Output:

Rahul -> 82.67

Anu -> 92.00

Kiran -> 83.67

```python
for name in stud_total_marks:
  print(f'{name} ->{stud_total_marks[name]/len(set(sub_name)):.2f}')
```

##3. Find  Top Scorer

Expected Output:

Anu -> 276

```python
highest= max(stud_total_marks.values())
for name in stud_total_marks:
  if stud_total_marks[name] ==highest:
    print(f'{name}->{stud_total_marks[name]}')
```

##4. Find Subject Topper

Expected Output:

Math -> Anu (92)

Science -> Anu (95)

English -> Kiran (91)

```python
sub_wise_highest_marks={}
for sub in set(sub_name):
  sub_wise_highest_marks[sub]=0
for sub , mark in zip(sub_name , marks):
  if mark > sub_wise_highest_marks[sub]:
    sub_wise_highest_marks[sub]=mark
for sub in sub_wise_highest_marks:
  for name ,subb, mark in students:
    if sub ==subb and sub_wise_highest_marks[sub] == mark:
      print(f'{sub}->{name} score {mark}')

```

##5. Assign Grades

Rule:

Average >= 90  -> A

Average >= 80  -> B

Average >= 70  -> C

Else           -> D

### Expected Output:

Rahul -> B

Anu -> A

Kiran -> B

```python
grade={}
for stu in stud_total_marks:
  if (stud_total_marks[stu] / 3) >= 90:
    print(f'{stu}->{(stud_total_marks[stu] / 3):.2f} -> A')
    grade[stu]='A'
  elif (stud_total_marks[stu] / 3) >= 80:
    print(f'{stu}->{(stud_total_marks[stu] / 3):.2f} -> B')
    grade[stu]='B'
  elif (stud_total_marks[stu] / 3) >= 70:
    print(f'{stu}->{(stud_total_marks[stu] / 3):.2f} -> C')
    grade[stu]='C'
  else:
    print(f'{stu}->{(stud_total_marks[stu] / 3):.2f} -> D')
    grade[stu]='D'




```

```python
print('========== STUDENT REPORT ========== \n')
stud_total_marks # it contain all students total marks  like student name : total_marks
grade # it contains all students with grades
top_score =float('-inf')
for name , mark in stud_total_marks.items():
  print('\n',name)
  print(f'Toatal marks :{mark} \nAverage :{(mark / 3):.2f} \ngrade :{grade[name]}')
  if mark > top_score :
    top_score=mark
    topper_name =name

print(f'\ntop score:- \n{topper_name,top_score}')
print('\nSubject Toppers:')
sub_wise_highest_marks # it contains subject wise highest score .....
for sub in sub_wise_highest_marks:
  for name ,subb, mark in students:
    if sub ==subb and sub_wise_highest_marks[sub] == mark:
      print(f'\n{sub}->{name} score {mark}')
```
