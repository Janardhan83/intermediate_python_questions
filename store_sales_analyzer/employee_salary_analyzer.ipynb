# Project 4: Store Sales Analyzer 🛒

Dataset

sales = [
    ("Laptop", "Electronics", 2, 50000),

    ("Mobile", "Electronics", 5, 20000),

    ("Shirt", "Clothing", 10, 1000),
    
    ("Jeans", "Clothing", 6, 2000),
    
    ("Rice", "Grocery", 20, 50),
    
    ("Oil", "Grocery", 15, 150),
    
    ("Laptop", "Electronics", 1, 50000),
    
    ("Rice", "Grocery", 10, 50)]
    
Each tuple contains:

(product_name, category, quantity_sold, price_per_unit)

### preparing data

```python

sales = [
    ("Laptop", "Electronics", 2, 50000),
    ("Mobile", "Electronics", 5, 20000),
    ("Shirt", "Clothing", 10, 1000),
    ("Jeans", "Clothing", 6, 2000),
    ("Rice", "Grocery", 20, 50),
    ("Oil", "Grocery", 15, 150),
    ("Laptop", "Electronics", 1, 50000),
    ("Rice", "Grocery", 10, 50)
]
```

##1.Calculate total revenue generated by each product.

**Revenue = Quantity × Price**

Expected Output:

Laptop -> 150000

Mobile -> 100000

Shirt -> 10000

Jeans -> 12000

Rice -> 1500

Oil -> 2250

```python
each_product_revenue ={}
for prd_name , cat , qnty, price_unit in sales:
  if prd_name in each_product_revenue:
    each_product_revenue[prd_name] =each_product_revenue[prd_name] + ( qnty * price_unit)
  else:
    each_product_revenue[prd_name] = qnty * price_unit
print('total revenue generated by each product : \n')
for name , price in each_product_revenue.items():
  print(name,'->',price )
```


##Find the highest revenue product.

Expected Output:

Laptop -> 150000

```python
max_total = max(each_product_revenue.values())
for name in each_product_revenue:
  if each_product_revenue[name] == max_total:
    print(f'highest higest revenue product :{name} salary :{each_product_revenue[name]}')
```

##3.Calculate total quantity sold for each product.

Expected Output:

Laptop -> 3

Mobile -> 5

Shirt -> 10

Jeans -> 6

Rice -> 30

Oil -> 15

```python
total_qnty_sold_for_each_prd={}
for prd_name , cat , qnty, price_unit in sales:
  if prd_name in total_qnty_sold_for_each_prd:
    total_qnty_sold_for_each_prd[prd_name] =total_qnty_sold_for_each_prd[prd_name]+ qnty
  else:
    total_qnty_sold_for_each_prd[prd_name]= qnty
print('total quantity sold for each product: \n')
for name , qnty in total_qnty_sold_for_each_prd.items():
  print(f'{name}:-{qnty}')
```

##4.Find the best-selling product.

Definition:

Highest quantity sold

Expected Output:

Rice -> 30 units

```python
print('the best-selling product:-')
max_qnty = max(total_qnty_sold_for_each_prd.values())
for name in total_qnty_sold_for_each_prd:
  if total_qnty_sold_for_each_prd[name] == max_qnty:
    print(f'{name} : {total_qnty_sold_for_each_prd[name]}')
```

##5.Calculate category-wise revenue.

Expected Output:

Electronics -> 250000

Clothing -> 22000

Grocery -> 3750


```python
category_wise_revenue={}
for prd_name , cat , qnty, price_unit in sales:
  if cat in category_wise_revenue:
    category_wise_revenue[cat] =category_wise_revenue[cat]+ (qnty*price_unit)
  else:
    category_wise_revenue[cat]= qnty*price_unit
print('total category-wise revenue: \n')
for cat , revenue in category_wise_revenue.items():
  print(f'{cat}:-{revenue}')
```

##6.Assign Product Performance

Rules:

Revenue >= 100000 → Excellent

Revenue >= 50000 → Good

Revenue >= 10000 → Average

Else → Low
____________________
Expected Output:

Laptop -> Excellent

Mobile -> Excellent

Shirt -> Average

Jeans -> Average

Rice -> Low

Oil -> Low

```python
# each_product_revenue . it contain product name and total revenue
Performance_Level={}
for name , revenue in each_product_revenue.items():
  if each_product_revenue[name] >=100000:
    Performance_Level[name] = 'Excellent'
  elif each_product_revenue[name] >= 50000:
    Performance_Level[name]='Good'
  elif each_product_revenue[name] >= 10000:
    Performance_Level[name] ='Average'
  else:
    Performance_Level[name]='Low'

for name , level in Performance_Level.items():
  print(f'{name} : {level}')

```

## 7.Calculate contribution percentage.

Formula:

(Product Revenue / Total Revenue) × 100

Example:

Laptop -> 54.13%

Mobile -> 36.08%

...



```python
import math
total_revenue=sum(each_product_revenue.values())
contribution_percentage = {}
for prd , total in each_product_revenue.items():
  contribution_percentage[prd] =round(((total /total_revenue) * 100 ),2)
print('contribution percentage of each product :')
for name , per in contribution_percentage.items():
  print(name ,'->' ,per)
```

```python
print('========== STORE SALES REPORT ========== \n')
print('Product Summary:')
for prd_name , cat , qnty, price_unit in sales:
  print(prd_name)
  print(f'{prd_name} ={each_product_revenue[prd_name]}')
  print(f'Quantity Sold :{total_qnty_sold_for_each_prd[prd_name]}')
  print(f'Performance :{Performance_Level[prd_name]}')
  print(f'Contribution :{contribution_percentage[prd_name]}\n')

max_total = max(each_product_revenue.values())
for name in each_product_revenue:
  if each_product_revenue[name] == max_total:
    print(f'highest higest revenue product :{name} salary :{each_product_revenue[name]}\n')

print('the best-selling product:-')
max_qnty = max(total_qnty_sold_for_each_prd.values())
for name in total_qnty_sold_for_each_prd:
  if total_qnty_sold_for_each_prd[name] == max_qnty:
    print(f'{name} : {total_qnty_sold_for_each_prd[name]}')

print('\ntotal category-wise revenue:')
for cat , revenue in category_wise_revenue.items():
  print(f'{cat}:-{revenue}')



```
