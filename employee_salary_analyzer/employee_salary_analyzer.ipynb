# Project 3: Employee Salary Analyzer 💼

Dataset

employees = [
    ("Rahul", "IT", 50000),

    ("Anu", "HR", 45000),

    ("Kiran", "IT", 60000),
    
    ("Priya", "Finance", 55000),
    
    ("Rahul", "IT", 10000),      
    
    ("Anu", "HR", 5000),         
    
    ("Kiran", "IT", 15000),      
    
    ("Priya", "Finance", 8000)   
]

Each tuple contains:

(employee_name, department, salary)

### preparing data

```python

employees = [
    ("Rahul", "IT", 50000),
    ("Anu", "HR", 45000),
    ("Kiran", "IT", 60000),
    ("Priya", "Finance", 55000),
    ("Rahul", "IT", 10000),      # Bonus
    ("Anu", "HR", 5000),         # Bonus
    ("Kiran", "IT", 15000),      # Bonus
    ("Priya", "Finance", 8000)   # Bonus
]
# storing each one in separate list ......
employee_name=[]
department=[]
salary=[]
for emp , deprt , slry in employees:
  employee_name.append(emp),department.append(deprt),salary.append(slry)
print(employee_name)
print(department)
print(salary)
```

##1.Calculate total earnings of each employee.

Expected Output:

Rahul -> 60000

Anu -> 50000

Kiran -> 75000

Priya -> 63000

```python
total_earnings_of_emp={}
for name , slry in zip(employee_name,salary):
  if name in total_earnings_of_emp:
    total_earnings_of_emp[name]=total_earnings_of_emp[name]+slry
  else:
    total_earnings_of_emp[name]=slry
for name , slry in total_earnings_of_emp.items():
  print(f'{name} ->{slry}')
```


##2.Find the highest earning employee.

Expected Output:

Kiran -> 75000

```python
highest_earn=float('-inf')
for name , slry in total_earnings_of_emp.items():
  if slry >= highest_earn:
    enp =name
    highest_earn=slry
print(f'highest earning employee :{enp} salary :{highest_earn}')
```

##3.Calculate total salary paid by each department.

Expected Output:

IT -> 135000

HR -> 50000

Finance -> 63000

```python
department_total_paid_salary={}
for deprt , slry in zip(department,salary):
  if deprt in department_total_paid_salary:
    department_total_paid_salary[deprt] =department_total_paid_salary[deprt]+slry
  else:
    department_total_paid_salary[deprt] =slry
for depr , slry in department_total_paid_salary.items():
  print(f'{depr}:-{slry}')
```

##4.Calculate average earning per employee.

Expected Output:

Rahul -> 30000

Anu -> 25000

Kiran -> 37500

Priya -> 31500

```python
total_earnings_of_emp # it contain total salary of employee
no_of_jobs_emp={}
for name in employee_name:
  if name in no_of_jobs_emp:
    no_of_jobs_emp[name]=no_of_jobs_emp[name]+1
  else:
    no_of_jobs_emp[name]=1
for name , slry in total_earnings_of_emp.items():
  print(f'{name} ->{(slry /no_of_jobs_emp[name]):.2f}')
```

##5.Assign Performance Level

Rules:

Average >= 35000 → Excellent

Average >= 30000 → Good

Average >= 25000 → Average

Else → Needs Improvement

Expected Output:

Rahul -> Good

Anu -> Average

Kiran -> Excellent

Priya -> Good

```python
Performance_Level={}
for name , sal in total_earnings_of_emp.items():
  if (sal / no_of_jobs_emp[name]) >= 35000:
    Performance_Level[name] = 'Excellent'
  elif (sal / no_of_jobs_emp[name]) >= 30000:
    Performance_Level[name]='Good'
  elif (sal / no_of_jobs_emp[name]) >= 25000:
    Performance_Level[name] ='Average'
  else:
    Performance_Level[name]='Needs Improvement'

for name , level in Performance_Level.items():
  print(f'{name} : {level}')

```

```python
print('========== EMPLOYEE REPORT ========== \n')
emp_departments={}
for name , dep  in zip(employee_name,department):
  emp_departments[name] = dep
for name , dep in emp_departments.items():
  print('\n',name,'\n')
  print('department :-',dep)
  print('Total earn :',total_earnings_of_emp[name])
  print(f'Average :{total_earnings_of_emp[name]/no_of_jobs_emp[name]}')
  print(f'performance :{Performance_Level[name]}')
print('\nHighest Earner: ')
maxx = max(total_earnings_of_emp.values())
for name , tol in total_earnings_of_emp.items():
  if total_earnings_of_emp[name] == maxx:
    print( name , tol)
print('\nDepartment Salary Summary:\n')
for dep , sl in department_total_paid_salary.items():
  print(dep,'->',sl)
```
